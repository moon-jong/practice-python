```python
import pandas as pd

wine_data = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
#멀티 인덱싱 -> 대분류(나라), 소분류(지역별) 가장 점수가 높은 와인 출력 
wine_data.groupby(['country','province']).apply(lambda df:df.loc[df.points.idxmax()])
```

```python
#agg -> 다중으로 함수 입력
#groupby로 묶으면 묶은 변수 -> 인덱스화
wine_data.groupby(['country','province']).price.agg([len, min, max])
```

```python
#sorting
wine_data.sort_values(by = ['country', 'points'], ascending = False)
```
