```python
import pandas as pd

wine_file_path = 'winemag-data-130k-v2.csv'
wine_data = pd.read_csv(wine_file_path)

wine_data.head(100)
```

```python
#colmn 이름으로 잡기

print(wine_data['country'])
print(wine_data['points'])
```

```python
#pandas는 기본적으로 [열][헹]의 인덱싱 구조를 갖는다

first_country = wine_data['country'][0]
first_point = wine_data['points'][0]
print(first_country,first_point)
```

```python
#iloc & loc
#iloc은 기본적으로 숫자기반의 인덱싱 구조를 채택
print(wine_data.iloc[0:1,0:2]) #iloc & loc은 [행][열] 구조를 가짐
wine_data.iloc[0:1][0:2] #두 출력의 차이가 있다. [행,열]로 구분

```

```python
print(wine_data.iloc[[1,3,5,6,7,9]]) #iloc인덱스에 리스트를 집어넣어 원하는 값만 출력 가능
print(wine_data.iloc[[1,3,5,7,8],[6,8]]) #마찬가지로 [[행리스트],[열리스트]]
print(wine_data.iloc[-5,:-6])#뒤에서부터 값을 가져옴
```

```python
#컬럼 이름 기반 인덱스 호출 loc
print(wine_data.loc[:15,'country'])
column_name = ['country', 'points', 'taster_name']
wine_data.loc[:15,column_name]
```

```python
#isin함수
#특정 컬럼에서 특정 값만 빼올 수 있다.
print(wine_data.country.isin(['France','Italy'])) #T,F로 나타남
wine_data.loc[wine_data.country.isin(['France', 'Italy'])]#나라 이름에 프랑스와 이탈리아만 호출
```

```python
#조건 호출 점수가 80점 이상인 프랑스, 이탈리아 와인
wine_data.loc[
    (wine_data.points >= 80)
    &
    (wine_data.country.isin(['France', 'Italy']))
]
```
