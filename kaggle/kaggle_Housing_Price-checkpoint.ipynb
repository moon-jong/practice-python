```python
import pandas as pd

file_path = 'train.csv'
melb_data = pd.read_csv(file_path,index_col = 'Id')
```

```python
numerical_col = [col for col in melb_data.columns if melb_data[col].dtype in ['int64','float64']]
X_with_SalePrice = melb_data[numerical_col]
X = X_with_SalePrice.drop(['SalePrice'],axis =1)
y = melb_data.SalePrice
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size= 0.8,test_size = 0.2, random_state = 1)
X_train
```

```python
from sklearn.impute import SimpleImputer
simple_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(simple_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(simple_imputer.transform(X_valid))
imputed_X_train
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
def score_model(X_train, X_valid, y_train = y_train, y_valid = y_valid):
    model = RandomForestRegressor(n_estimators = 10, random_state=1)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid,preds)
```

```python
score_model(imputed_X_train,imputed_X_valid)
```

```python
test_data = pd.read_csv('test.csv',index_col = 'Id')
test_nummeric_data = [col for col in test_data.columns if test_data[col].dtype in ['int64', 'float64']]
```

```python
test_X = test_data[test_nummeric_data]
test_X
```

```python
test_imputer = SimpleImputer()

imputed_text_X = pd.DataFrame(test_imputer.fit_transform(test_X))
imputed_text_X
```

```python
model = RandomForestRegressor(n_estimators=10 , random_state=1)
model.fit(imputed_X_train,y_train)
```

```python
predict = model.predict(imputed_text_X)
```

```python
output = pd.DataFrame({'Id':test_X.index,
                        'SalePrice':predict})
```
