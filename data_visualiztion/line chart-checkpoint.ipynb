# Line chart Drawing with seaborn
    ## import package : pandas, matplotlib, seaborn

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup complete")
```

- call the data

```python
#state the file path
spotify_filepath = "spotify.csv"

#call the csv file to an object
spotify_data = pd.read_csv(spotify_filepath, index_col = "Date", parse_dates = True)

#check
spotify_data.tail()
```

 - print line chart by using seaborn module

```python
#configurate the fig size
plt.figure(figsize = (16,9))

#set the title
plt.title("Most frequent streamed Music in 2018")

#draw chart
sns.lineplot(data = spotify_data)
```

```python
#configure size
plt.figure(figsize = (16,9))

#print speratly
sns.lineplot(data = spotify_data["HUMBLE."], label = "HUMBLE", color = 'r')

#set x_label
plt.xlabel("Date")
```
