# Print Barchart & Heatmap by using seaborn

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns

print("Setup complete")
```

```python
#define the file path (in this example will use flight_delays.csv)
flight_delay_filepath = "flight_delays.csv" # it's in the workspace when it doesn't exist at the wokrspace then "../../flight_delays.csv"

#read csv data
flight_delay_data = pd.read_csv(flight_delay_filepath, index_col = "Month")

#check
flight_delay_data
```

```python
#set the figure_Size
plt.figure(figsize = (10,6))

#print the barplot
sns.barplot(x = flight_delay_data.index, y = flight_delay_data["VX"])
plt.title("Month average Delay for VX Airline")
plt.xlabel("Month")
```

```python
# set the figsize
plt.figure(figsize = (10,7))

#print the Heatmap
plt.title("Average Arrival Delay")
sns.heatmap(flight_delay_data, annot= True)
```
