```python
#visualize ign_scores

import pandas as pd
pd.plotting.deregister_matplotlib_converters()

import matplotlib.pyplot as plt
import seaborn as sns

print("setup is completed")
```

```python
#set the path object
ign_filepath = "ign_scores.csv"

ign_data = pd.read_csv(ign_filepath, index_col ="Platform")
ign_data
```

```python
#print barplot
plt.figure(figsize = (16,9))
plt.title("The Shooter genre average score by Platform")
sns.barplot(x = ign_data.index, y = ign_data["Shooter"])
plt.xlabel("Platform")
plt.xticks(rotation = -45)
```

```python
plt.figure(figsize = (16,9))
plt.title("the Average value of the IGN score for each Platform and Genre")
sns.heatmap(ign_data, annot = True)
```

```python
#scatter plot
plt.figure(figsize = (16,9))
plt.title("scatter plot for score of Play station4")
plt.xticks(rotation = -45)
sns.scatterplot(x= ign_data.index, y = ign_data["Fighting"])
```
